# 01. Data Collection

Fetch sector ETFs (XLF, XLK, XLE, XLU, XLY, XLP, XLV, XLI) and risk factors (VIX, 10-year yield, 3-month yield) from Yahoo Finance.

The goal is to build a dataset for predicting market stress regimes. Stress periods are typically characterized by rising VIX, falling yields, and sector outflows.

```python
import sys
sys.path.insert(0, '..')

from data_fetch import fetch_prices, save_raw, TICKERS, START_DATE, END_DATE

print(f"Fetching data from {START_DATE} to {END_DATE}")
print(f"Tickers: {TICKERS}")
```

## Download prices

```python
prices = fetch_prices(TICKERS, START_DATE, END_DATE)
print(f"\nData shape: {prices.shape[0]} rows (trading days), {prices.shape[1]} tickers")
print(f"Date range: {prices.index.min().date()} to {prices.index.max().date()}")
```

## Check data quality

```python
# Missing data
missing = prices.isna().sum()
print("Missing values per ticker:")
print(missing[missing > 0].sort_values(ascending=False))

if missing.sum() == 0:
    print("\nNo missing values.")
```

## Summary statistics

```python
print("\nFirst 5 rows:")
print(prices.head())

print("\nBasic statistics:")
print(prices.describe())
```

## Save raw data

```python
save_raw(prices, '../data/raw/prices.csv')
print(f"Saved to data/raw/prices.csv")
```
