# 01. Data Collection

Fetch eight sector ETFs and three market risk series from Yahoo Finance. The download is pinned
through 10 July 2026 so the downstream results have a fixed end date. Yahoo treats the `end`
argument as exclusive, so the source module requests 11 July.

```python
import sys
sys.path.insert(0, '..')

from src.data_fetch import fetch_prices, save_raw, TICKERS, START_DATE, END_DATE

print(f"Fetching data from {START_DATE} to {END_DATE}")
print(f"Tickers: {TICKERS}")
```

## Download prices

```python
prices = fetch_prices(TICKERS, START_DATE, END_DATE)
print(f"\nData shape: {prices.shape[0]} rows (trading days), {prices.shape[1]} tickers")
print(f"Date range: {prices.index.min().date()} to {prices.index.max().date()}")
```

## Check data quality

```python
# Missing data
missing = prices.isna().sum()
print("Missing values per ticker:")
print(missing[missing > 0].sort_values(ascending=False))

if missing.sum() == 0:
    print("\nNo missing values.")
```

## Summary statistics

```python
print("\nFirst 5 rows:")
print(prices.head())

print("\nBasic statistics:")
print(prices.describe())
```

## Save raw data

```python
save_raw(prices, '../data/raw/prices.csv')
print(f"Saved to data/raw/prices.csv")
```
